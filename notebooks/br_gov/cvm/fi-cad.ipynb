# Fundos de Investimento: Informação Cadastral

Dados cadastrais de fundos de investimento estruturados e não estruturados (ICVM 555), tais como: CNPJ, data de registro e situação do fundo.

O conjunto de dados disponibiliza:

- Fundos de Investimento - Conjunto Completo: As informações cadastrais referentes ao último dia útil.
- Fundos de Investimento ICVM 555 - Histórico: O histórico de alteração de informações cadastrais (atualização diária).

A atualização dos arquivos ocorre de terça a sábado, às 08:00h, com a posição cadastral dos fundos até as 23:59h do dia anterior.

```python
import pandas as pd
import requests
from io import BytesIO
```

```python
def get_cad_fi() -> pd.DataFrame:
    url = "https://dados.cvm.gov.br/dados/FI/CAD/DADOS/cad_fi.csv"
    data = requests.get(url).content
    df = pd.read_csv(BytesIO(data), sep=";", encoding="ISO-8859-1")
    return df

df = get_cad_fi()
df
```

```python
df2 = df.groupby("CNPJ_FUNDO").last().reset_index(drop=True)
len(df2)
```

```python
from zipfile import ZipFile


def get_fi_cad_historical():
    url = "https://dados.cvm.gov.br/dados/FI/CAD/DADOS/cad_fi_hist.zip"
    data = requests.get(url).content
    with ZipFile(BytesIO(data)) as z:
        for f in z.namelist():
            if f.endswith(".csv"):
                with z.open(f) as zf:
                    df = pd.read_csv(zf, sep=";", encoding="ISO-8859-1")
                    df.Name = f[:-4]
                    col_names = {}
                    for col in df.columns:
                        col_names[col] = col.lower()
                    df.rename(columns=col_names, inplace=True)
                    yield df


hist = list(get_fi_cad_historical())
for df in hist:
    print(df.Name, len(df))
```
